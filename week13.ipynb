```python
# 실습#1 생일 기온 시각화
#202012316 이지영
import csv
import matplotlib.pyplot as plt
f=open('seoul_10.csv')
data=csv.reader(f)
header=next(data)
tem_high=[]
tem_low=[]
for row in data :
    month=row[0].split('-')[1]
    day=row[0].split('-')[2]
    if row[-1] !='':
        if month=='03' and day=='09':
            tem_high.append(float(row[-1]))
            tem_low.append(float(row[-2]))
f.close()
plt.rc('font', family='Malgun Gothic')
plt.title('최근 10년간 3월 9일의 최고기온과 최저기온 그래프\n 202012316 이지영')
plt.plot(tem_high, color='pink', label='최고기온')
plt.plot(tem_low, color='skyblue', label='최저기온')
plt.legend()
plt.show()
```

```python
# 실습#2 최고 기온 시각화
#202012316 이지영
import csv
import matplotlib.pyplot as plt
f=open('seoul_10.csv')
data=csv.reader(f)
header=next(data)
aug=[]
jan=[]
for row in data :
    month=row[0].split('-')[1]
    if row[-1] !='':
        if month=='01':
            jan.append(float(row[-1]))
        if month=='08':
            aug.append(float(row[-1]))
        
f.close()
plt.rc('font', family='Malgun Gothic')
plt.title('최근 10년간 서울 지역의 1월과 8월 최고 기온 데이터\n 202012316 이지영')
plt.hist(jan, bins=100, density = True, 
         histtype ='bar',color='b', label='Jan')
plt.hist(aug, bins=100, density = True, 
         histtype ='bar',color='r', label='Aug')
plt.legend()
plt.show()
```
