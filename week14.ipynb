```python
# 실습#1 인구 구조
#202012316 이지영
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mn=10000000000000
mx=-100
mn_name=""
mx_name=""
n_result=[]
x_result=[]

###############

f=open('202012_202012_연령별인구현황_월간.csv')
data=csv.reader(f)
header=next(data)

for row in data :
    if row[0] !=' ':
        if row[0] =='광진구':
            name_1=row[0]
            home=np.array(row[3:], dtype=int)
                #home에 광진구의 인구 비율 저장
f.close()    

###############비슷한

f=open('202012_202012_연령별인구현황_월간.csv')
data=csv.reader(f)
header=next(data)
for row in data : 
    if row[0] !='광진구'and row[0] !='서울특별시':
        for i in range(3,104):
            row[i]=int(row[i])
        away=np.array(row[3:], dtype=int)
        s=np.sum((away-home)**2)
        if s<mn:
            mn=s
            mn_name=row[0]  
            n_result=away
f.close()

###############다른

f=open('202012_202012_연령별인구현황_월간.csv')
data=csv.reader(f)
header=next(data)
for row in data : 
    if row[0] !='광진구'and row[0] !='서울특별시':
        for i in range(3,104):
            row[i]=int(row[i])
        away=np.array(row[3:], dtype=int)
        s=np.sum((home-away)**2)
        if s>mx:
            mx=s
            mx_name=row[0]
            x_result=away
f.close()        
###############

plt.rc('font', family='Malgun Gothic')
plt.title('광진구와 가장 비슷한 인구 구조를 가진 지역\n 202012316 이지영')
plt.style.use('ggplot')
plt.plot(home, color='red', label=name_1)
plt.plot(n_result, color='darkgreen', label=mn_name+"(표준편차 : "+str((mn/101)**0.5)+")")
plt.legend()
plt.show()

plt.rc('font', family='Malgun Gothic')
plt.title('광진구와 가장 다른 인구 구조를 가진 지역\n 202012316 이지영')
plt.style.use('ggplot')
plt.plot(home, color='red', label=name_1)
plt.plot(x_result, color='darkturquoise', label=mx_name+"(표준편차 : "+str((mx/101)**0.5)+")")
plt.legend()
plt.show()
```
