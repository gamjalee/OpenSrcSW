```python
# 실습#1 인구 구조
#202012316 이지영
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mn=1
mx=0.0003
#mn_name=[' ']*3
mn_name=[]
mx_name=[]
n_result=[]
x_result=[]
n_stdeviation=[]
x_stdeviation=[]
i=0
j=0
###############
f=open('202012_202012_연령별인구현황_월간.csv')
data=csv.reader(f)
header=next(data)

for row in data :
    if row[0] !=' ':
        if row[0] =='광진구':
            name_1=row[0]
            home=np.array(row[3:], dtype=int)/int(row[2])
                #home에 광진구의 인구 비율 저장
f.close()    
###############비슷한
f=open('202012_202012_연령별인구현황_월간.csv')
data=csv.reader(f)
header=next(data)

for row in data : 
    away=np.array(row[3:], dtype=int)/int(row[2])
    s=np.sum((home-away)**2)
    if s<mn and row[0] !='광진구'and row[0] !='서울특별시':
        mn=s
        mn_name.append(row[0])
        #print(mn_name)
        n_result.append(away)
        #print(n_result)
        n_stdeviation.append((s/int(row[2]))**0.5)
        #print(n_stdeviation)
f.close()
###############다른
f=open('202012_202012_연령별인구현황_월간.csv')
data=csv.reader(f)
header=next(data)

for row in data : 
    away=np.array(row[3:], dtype=int)/int(row[2])
    s=np.sum((home-away)**2)
    if s>mx and row[0] !='광진구' :
        mx=s
        mx_name.append(row[0])
        #print(mx_name)
        x_result.append(away)
        #print(x_result)
        x_stdeviation.append((s/int(row[2]))**0.5)
        #print(x_stdeviation)
f.close()        
###############

# print("mn")
# print(mn_name)
# #print(n_result)
# print(n_stdeviation)
# print("mx")
# print(mx_name)
# #print(x_result)
# print(x_stdeviation)

plt.rc('font', family='Malgun Gothic')
plt.title('광진구와 가장 비슷한 인구 구조를 가진 지역\n 202012316 이지영')
plt.style.use('ggplot')
plt.plot(home, color='red', label=name_1)
plt.plot(n_result[0], color='limegreen', label=mn_name[0]+"(표준편차 : "+str(n_stdeviation[0])+")")
plt.plot(n_result[1], color='lime', label=mn_name[1]+"(표준편차 : "+str(n_stdeviation[1])+")")
plt.plot(n_result[2], color='darkgreen', label=mn_name[2]+"(표준편차 : "+str(n_stdeviation[2])+")")
plt.legend()
plt.show()

plt.rc('font', family='Malgun Gothic')
plt.title('광진구와 가장 다른 인구 구조를 가진 지역\n 202012316 이지영')
plt.style.use('ggplot')
plt.plot(home, color='red', label=name_1)
plt.plot(x_result[0], color='cyan', label=mx_name[0]+"(표준편차 : "+str(x_stdeviation[0])+")")
plt.plot(x_result[1], color='darkturquoise', label=mx_name[1]+"(표준편차 : "+str(x_stdeviation[1])+")")
plt.plot(x_result[2], color='teal', label=mx_name[2]+"(표준편차 : "+str(x_stdeviation[2])+")")
plt.legend()
plt.show()
```
