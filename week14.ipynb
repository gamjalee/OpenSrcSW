```python
# #1  
#202012316 
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mn=10000000000000
mx=-100
mn_name=""
mx_name=""
n_result=[]
x_result=[]

###############

f=open('202012_202012__.csv')
data=csv.reader(f)
header=next(data)

for row in data :
    if row[0] !=' ':
        if row[0] =='':
            name_1=row[0]
            home=np.array(row[3:], dtype=int)
                #home    
f.close()    

###############

f=open('202012_202012__.csv')
data=csv.reader(f)
header=next(data)
for row in data : 
    if row[0] !=''and row[0] !='':
        for i in range(3,104):
            row[i]=int(row[i])
        away=np.array(row[3:], dtype=int)
        s=np.sum((away-home)**2)
        if s<mn:
            mn=s
            mn_name=row[0]  
            n_result=away
f.close()

###############

f=open('202012_202012__.csv')
data=csv.reader(f)
header=next(data)
for row in data : 
    if row[0] !=''and row[0] !='':
        for i in range(3,104):
            row[i]=int(row[i])
        away=np.array(row[3:], dtype=int)
        s=np.sum((home-away)**2)
        if s>mx:
            mx=s
            mx_name=row[0]
            x_result=away
f.close()        
###############

plt.rc('font', family='Malgun Gothic')
plt.title('      \n 202012316 ')
plt.style.use('ggplot')
plt.plot(home, color='red', label=name_1)
plt.plot(n_result, color='darkgreen', label=mn_name+"( : "+str((mn/101)**0.5)+")")
plt.legend()
plt.show()

plt.rc('font', family='Malgun Gothic')
plt.title('      \n 202012316 ')
plt.style.use('ggplot')
plt.plot(home, color='red', label=name_1)
plt.plot(x_result, color='darkturquoise', label=mx_name+"( : "+str((mx/101)**0.5)+")")
plt.legend()
plt.show()
```
